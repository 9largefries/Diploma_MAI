## EDA - Exploratory Data Analysis

#### Импорт библиотек

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Чтение данных

```python
df = pd.read_parquet('data/train_data.parquet')
```

```python
cols = ['setting1', 'setting2', 'setting3', 's1', 's2', 's3',
       's4', 's5', 's6', 's7', 's8', 's9', 's10', 's11', 's12', 's13', 's14',
       's15', 's16', 's17', 's18', 's19', 's20', 's21']
```

#### График значений сенсоров двигателя 1

```python
df[df.id == 1][cols].plot(subplots=True, figsize=(15,30))
plt.show()
```

#### Рассмотрим матрицу корреляций признаков

```python
corr = df[cols].corr()
corr.style.background_gradient(cmap='coolwarm')
```

#### Рассмотрим корреляцию признаков с значением RUL

```python
df[cols].corrwith(df.RUL).abs().sort_values(ascending=False).plot.bar()
plt.show()
```
